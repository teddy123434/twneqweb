```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import pandas as pd

def fetch_earthquake_data(page):
    url = "https://scweb.cwa.gov.tw/zh-tw/earthquake/ajaxhandler"  # 替換為實際的 AJAX URL
    headers = {
        "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0"
    }
    params = {
        "page": page  # 替換為實際的參數
    }
    response = requests.get(url, headers=headers, params=params)
    data = response.json()  # 假設回應是 JSON 格式

    # 解析數據 (需要根據實際數據結構調整)
    for item in data:
        id = item['id']
        time = item['time']
        magnitude = item['magnitude']
        depth = item['depth']
        apxlocation = item['apxlocation']

        print(f"[編號] {id} [日期時間] {time} [芮氏規模] {magnitude} [深度] {depth} [大略位置] {apxlocation}")

# 獲取前10頁的數據
for page in range(1, 11):
    fetch_earthquake_data(page)
```
